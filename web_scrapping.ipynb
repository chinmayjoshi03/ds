```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://books.toscrape.com/catalogue/page-1.html"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

data = []

books = soup.find_all("article", class_="product_pod")

for book in books[:5]:
    title = book.h3.a["title"]
    rating = book.p["class"][1]

    link = "https://books.toscrape.com/catalogue/" + book.h3.a["href"].replace("../", "")


    res2 = requests.get(link)
    soup2 = BeautifulSoup(res2.text, "html.parser")

    desc = soup2.find("meta", attrs={"name": "description"})
    review = desc["content"].strip() if desc else "No review"

    data.append([title, rating, review])


df = pd.DataFrame(data, columns=["Product", "Rating", "Review"])
df.to_csv("output.csv", index=False)

print(df)
```
