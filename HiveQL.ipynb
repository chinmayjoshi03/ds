```python
'''
✅ STEP 1: Open Hive

In Cloudera terminal:

hive
✅ STEP 2: Create Database
CREATE DATABASE student_db;
SHOW DATABASES;
USE student_db;
✅ STEP 3: Create Table (Student Record)
CREATE TABLE student(
    id INT,
    name STRING,
    age INT,
    marks FLOAT
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ',';

Check table:

SHOW TABLES;
✅ STEP 4: Load Data into Table

👉 First create a file in Linux:

nano student.txt

Add sample data:

1,Rahul,20,85
2,Priya,21,90
3,Amit,19,78
4,Sneha,22,88

Save file.

👉 Load into Hive:

LOAD DATA LOCAL INPATH '/home/cloudera/student.txt'
INTO TABLE student;

Check data:

SELECT * FROM student;
✅ STEP 5: Alter Table

👉 Add new column:

ALTER TABLE student ADD COLUMNS (city STRING);

👉 Rename table:

ALTER TABLE student RENAME TO student_info;

👉 View structure:

DESCRIBE student_info;
✅ STEP 6: Create External Table
CREATE EXTERNAL TABLE student_ext(
    id INT,
    name STRING,
    age INT,
    marks FLOAT
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ','
LOCATION '/user/cloudera/student_data';
✅ STEP 7: Load Data into External Table

👉 Move file to HDFS:

hdfs dfs -mkdir /user/cloudera/student_data
hdfs dfs -put student.txt /user/cloudera/student_data

👉 Now check:

SELECT * FROM student_ext;
✅ STEP 8: Drop Table / Database
DROP TABLE student_info;
DROP DATABASE student_db;

(If error comes → use CASCADE)

DROP DATABASE student_db CASCADE;

✅ C. Insert New Values and Add New Field
🔹 1) Insert New Values into Table

👉 Add a new student record:

INSERT INTO TABLE student_info
VALUES (5, 'Riya', 20, 92);

👉 Check:

SELECT * FROM student_info;
🔹 2) Insert Multiple Rows
INSERT INTO student_info VALUES
(6, 'Karan', 21, 80),
(7, 'Neha', 22, 85);
🔹 3) Add New Field (Column)

👉 Add column city:

ALTER TABLE student_info
ADD COLUMNS (city STRING);

👉 Check structure:

DESCRIBE student_info;
🔹 4) Update Field (Important Note ⚠️)

Hive does NOT support direct UPDATE like MySQL (in basic setup)

👉 So say in exam:

“Hive uses INSERT OVERWRITE instead of UPDATE”

Example:

INSERT OVERWRITE TABLE student_info
SELECT id, name, age, marks, 'Pune'
FROM student_info;
✅ D. Join Tables in Hive
🔹 Step 1: Create Second Table
CREATE TABLE course(
    id INT,
    course_name STRING
)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY ',';
🔹 Step 2: Insert Data
INSERT INTO course VALUES
(1, 'IT'),
(2, 'CS'),
(3, 'ENTC'),
(5, 'IT');
🔹 Step 3: Perform JOIN
✔ INNER JOIN
SELECT s.id, s.name, s.marks, c.course_name
FROM student_info s
JOIN course c
ON s.id = c.id;
✔ LEFT JOIN
SELECT s.id, s.name, c.course_name
FROM student_info s
LEFT JOIN course c
ON s.id = c.id;
✔ RIGHT JOIN
SELECT s.id, s.name, c.course_name
FROM student_info s
RIGHT JOIN course c
ON s.id = c.id;
'''
```
