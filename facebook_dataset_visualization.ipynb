```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/dataset_Facebook.csv", delimiter=';')
df.head()
```

```python
sns.boxplot(x='Type', y='like', data=df)
plt.title("Boxplot")
plt.show()
```

```python
plt.hist(df['Total Interactions'].dropna())
plt.title("Histogram")
plt.show()
```

```python
plt.plot(df['Total Interactions'])
plt.title("Line Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['like'],    label='Likes',    color='blue')
plt.plot(df.index, df['share'],   label='Shares',   color='green')
plt.plot(df.index, df['comment'], label='Comments', color='red')
plt.title("Multiple Line Graph - Likes, Shares & Comments per Post")
plt.xlabel("Post Index")
plt.ylabel("Count")
plt.legend()
plt.show()
```
